```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
```

```python
rdata = pd.read_csv('classification_train.csv')
rdata = pd.DataFrame(rdata)

data = np.zeros((int(rdata.shape[0]*0.8), 786))
data = pd.DataFrame(data)
for i in range(int(rdata.shape[0]*0.8)):
    data.iloc[i,:] = rdata.iloc[i,:]

label = data.iloc[:,1]
n_label = len(label.unique())
label = pd.get_dummies(label)
data.drop(data.columns[[0,1]], axis = 1, inplace = True)
features = data
features = pd.DataFrame(features)
```

```python
n_samples, n_features = features.shape
weight = np.random.rand(n_features,n_label)
weight = pd.DataFrame(weight)
bias = np.zeros(10)
z = np.zeros((label.shape[0],10))
z = pd.DataFrame(z)
lr = .8
cost = []
epoch = 20
#print(weight[0])
```

```python
# training the model
for k in range(epoch):
    
    lcost = []
    dm = lr*((np.dot(features.T, z - label))/features.shape[0])
    db = lr*((z - label).sum())
    weight = weight - dm
    bias = bias - db
    
    cost.append((((z - label)**2).sum())/(n_samples*2))
    
    for i in range(10):
        z[i] = np.dot(features, weight[i]) + bias[i]  
    z1 = z.T.sum()
    
    for j in range(z.shape[0]):
        z.iloc[j,:] = z.iloc[j,:]/z1[j]
    
    print(k)
```

```python
# making the data from train data for testing
tdata = np.zeros((rdata.shape[0] - int(rdata.shape[0]*0.8), 786))
tdata = pd.DataFrame(tdata)
for i in range(rdata.shape[0] - int(rdata.shape[0]*0.8)):
    tdata.iloc[i,:] = rdata.iloc[i,:]
    
tlabel = tdata.iloc[:,1]
tdata.drop(tdata.columns[[0,1]], axis = 1, inplace = True) 
```

```python
#finding the predicted data on the updated weight and bias
yt = np.dot(tdata, weight)
for m in range(weight.shape[1]): 
    for l in range(tdata.shape[0]):
        yt[l][m] = yt[l][m] + bias.T[m]
        
yt = pd.DataFrame(yt)
for j in range(tdata.shape[0]):
    yt.iloc[j,:] = yt.iloc[j,:]/yt.T.sum()[j]
```

```python
#testing the accuracy
a = 0
for i in range(6000):
    if yt.idxmax(axis = 1)[i] == tlabel[i]:
        a = a + 1
print(a/6000)
```

```python
bias = pd.DataFrame(bias)
test = pd.read_csv('classification_test.csv')
test = test.drop(['ids'], axis = 1)
zval = np.dot(test, weight)
for m in range(weight.shape[1]): 
    for l in range(test.shape[0]):
        zval[l][m] = zval[l][m] + bias.T[m]
```

```python
zval = pd.DataFrame(zval)
zval1 = zval.T.sum()
for j in range(test.shape[0]):
    zval.iloc[j,:] = zval.iloc[j,:]/zval1[j]
print(zval.idxmax(axis = 1))
```
