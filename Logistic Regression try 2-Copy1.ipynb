```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
```

```python
rdata = pd.read_csv('classification_train.csv')
rdata = pd.DataFrame(rdata)

data = np.zeros((int(rdata.shape[0]*0.8), 786))
data = pd.DataFrame(data)
for i in range(int(rdata.shape[0]*0.8)):
    data.iloc[i,:] = rdata.iloc[i,:]

label = data.iloc[:,1]
n_label = len(label.unique())
label = pd.get_dummies(label)
data.drop(data.columns[[0,1]], axis = 1, inplace = True)
features = data
features = pd.DataFrame(features)
```

```python
n_samples, n_features = features.shape
weight = np.random.randn(n_features,n_label)
weight = pd.DataFrame(weight)
bias = np.random.randn(10)
lr = .006
epoch = 250
cost = []
yp = []
#print(weight[0])
```

```python
for i in range(len(label.columns)):
    for j in range(epoch):
        z = np.dot(features,weight[i]) + bias[i]
        y_pred = 1/(1+(np.exp(-z)))
        dm = lr*((np.dot(features.T, y_pred - label[i]))/(features.shape[0]))
        db =  lr*(np.mean(y_pred - label[i]))
        weight[i] = weight[i] - dm
        bias[i] = bias[i] - db
        
#cost.append(np.mean(-label[i]*(np.log(y_pred)) - (1 - label[i])*np.log(1-y_pred))) 
```

```python
# making the data from train data for testing
tdata = np.zeros((rdata.shape[0] - int(rdata.shape[0]*0.8), 786))
tdata = pd.DataFrame(tdata)
for i in range(rdata.shape[0] - int(rdata.shape[0]*0.8)):
    tdata.iloc[i,:] = rdata.iloc[i,:]
    
tlabel = tdata.iloc[:,1]
tdata.drop(tdata.columns[[0,1]], axis = 1, inplace = True) 
```

```python
#finding the predicted data on the updated weight and bias
yt = np.dot(tdata, weight)
for m in range(weight.shape[1]): 
    for l in range(tdata.shape[0]):
        yt[l][m] = yt[l][m] + bias.T[m]
```

```python
#testing the accuracy
yt = pd.DataFrame(yt)
a = 0
for i in range(6000):
    if yt.idxmax(axis = 1)[i] == tlabel[i]:
        a = a + 1
print(a/6000)
```
